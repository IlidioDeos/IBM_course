```python
# TASK 3: Calculate the distances between a launch site to its proximities
import folium
import pandas as pd

# Suponha que essas são as coordenadas do local de lançamento e algumas coordenadas de pontos de proximidade
launch_site_coords = (29.559684888503615, -95.0830971930759)
proximity_coords = [
    (30.2672, -97.7431),  # Coordenadas de um ponto de proximidade (exemplo: cidade)
    (31.5493, -97.1467),  # Coordenadas de outro ponto de proximidade (exemplo: outra cidade)
    (29.7604, -95.3698)   # Coordenadas de outro ponto de proximidade (exemplo: terceira cidade)
]

# Criar um DataFrame com os pontos de proximidade
proximity_df = pd.DataFrame(proximity_coords, columns=['Lat', 'Lon'])

# Adicionar uma coluna 'Distance' ao DataFrame para armazenar as distâncias calculadas
proximity_df['Distance'] = proximity_df.apply(
    lambda row: ((launch_site_coords[0] - row['Lat'])**2 + (launch_site_coords[1] - row['Lon'])**2)**0.5, 
    axis=1
)

# Exibir o DataFrame com as distâncias calculadas
print(proximity_df)

# Criar um mapa
site_map = folium.Map(location=launch_site_coords, zoom_start=6)

# Adicionar marcador para o local de lançamento
folium.Marker(location=launch_site_coords, tooltip='Local de Lançamento').add_to(site_map)

# Adicionar marcadores para os pontos de proximidade
for index, row in proximity_df.iterrows():
    folium.Marker(location=(row['Lat'], row['Lon']), tooltip=f'Distância: {row["Distance"]:.2f}').add_to(site_map)

# Mostrar o mapa
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
